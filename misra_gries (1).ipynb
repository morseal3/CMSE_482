```python
import pandas as pd
from collections import Counter, OrderedDict
column_names = ['RevisionId', 'ArticleName', 'Timestamp', 'Editor']
data = pd.read_table('wiki_edit2.txt',sep=' ',header=None,names=column_names)
data
```

```python
d = []
for i in data['ArticleName']:
    d.append([i.strip()])
stream = pd.DataFrame(d, columns= ['ArticleName'])
stream
```

```python
 f = list(data['ArticleName'])
f
```

```python
def misra_gries(d_frame,k):
    f = list(d_frame['ArticleName'])
    counter = Counter()
    ## k is 1000
    for token in f:
        #add to frequency
        if token in f:
            counter[token] += 1
        elif len(counter) < k-1:
            counter[token] = 1
        else:
            for art_name in list(counter.keys()): 
                counter[art_name] -= 1
            if counter[art_name] == 0:
                del counter[art_name]
    result = OrderedDict(counter.most_common(1000))
    return result
        
        
```

```python
mg = misra_gries(stream,1000)
```
