```python
def byUnicode(char):
    if ord(char) < 48 or ord(char) > 127: return ''
    else: return char

try:
    cnx = mysql.connector.connect(user=uname, password=pwd, host=hname, database=uname)
    cursor = cnx.cursor()
    
    for row in bb_data:
        User = repr(''.join(list(filter(byUnicode, row['User']))[:50]))
        SKU = row['SKU']
        Category = repr(''.join(list(filter(byUnicode, row['Category']))[:50]))
        Query = repr(''.join(list(filter(byUnicode, row['Query']))[:100]))
        Click_time = repr(''.join(list(filter(byUnicode, row['ClickTime']))[:50]))
        Query_time = repr(''.join(list(filter(byUnicode, row['QueryTime']))[:50]))
        
        
        query = "INSERT INTO bestbuy VALUES ({}, {}, {}, {}, {}, {})"\
        .format(User, SKU, Category, Query, Click_time, Query_time)
        cursor.execute(query)
        
        
    cnx.commit()
    cursor.close()
    cnx.close()
    
except mysql.connector.Error as error:
    print(error)   
```
